```python
menu_items = {
    "Pizza": 200,
    "Burger": 60,
    "Pasta": 50,
    "Momos": 80,
    "Coffee": 40,
    "Tea": 20
}

print("\nðŸŒŸ Welcome to Cloud Kitchen! ðŸŒŸ\n")
print("Here's our menu:")
print("-" * 40)
for item, price in menu_items.items():
    print(f"{item:<10} - â‚¹{price}")
print("-" * 40)

cart = {}  
subtotal = 0
delivery_fee = 40
discount_threshold = 500
discount_rate = 0.10

while True:
    user_choice = input("\nEnter the item you'd like to order (or type 'done' to finish): ").capitalize()
    if user_choice.lower() == 'done':
        break
    elif user_choice in menu_items:
        try:
            quantity = int(input(f"How many {user_choice}s would you like to order? "))
            if quantity <= 0:
                print(" Please enter a valid quantity.")
                continue

            # Update cart and calculate costs
            if user_choice in cart:
                cart[user_choice] += quantity
            else:
                cart[user_choice] = quantity

            subtotal += menu_items[user_choice] * quantity
            print(f"âœ… Added {quantity} {user_choice}(s) to your cart.")
        except ValueError:
            print(" Please enter a valid number for quantity.")
    else:
        print(f" Sorry, we don't have {user_choice} on the menu.")

# Check if cart is empty
if not cart:
    print("\n No items were ordered. Please visit us again!")
else:
    print("\nðŸ§¾ Order Summary:")
    print("-" * 40)
    for item, quantity in cart.items():
        item_cost = menu_items[item] * quantity
        print(f"{item:<10} x {quantity:<3} = â‚¹{item_cost}")
    print("-" * 40)
    print(f"Subtotal: â‚¹{subtotal:.2f}")

    # Apply discount if applicable
    if subtotal >= discount_threshold:
        discount = subtotal * discount_rate
        print(f"ðŸŽ‰ Discount Applied: -â‚¹{discount:.2f} (10% off on orders â‚¹500 and above)")
        subtotal -= discount
    elif subtotal < 300:
        print(f"ðŸšš Delivery Fee: +â‚¹{delivery_fee}")
        subtotal += delivery_fee

    print("-" * 40)
    print(f"Total Amount to Pay: â‚¹{subtotal:.2f}")
    print("\nðŸ’– Thank you for ordering from Cloud Kitchen! Have a great day! ðŸŒŸ")
```
